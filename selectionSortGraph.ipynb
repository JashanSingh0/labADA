```python
import matplotlib.pyplot as plt

values = [
    (500, 0),
    (2000, 4),
    (3500, 12),
    (5000, 25),
    (6500, 43),
    (8000, 63),
    (9500, 87)
]

x_values, y_values = zip(*values)

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')

plt.xlabel('Input Values')
plt.ylabel('Seconds')
plt.title('Inputs Vs Time Graph')

plt.show()
```
